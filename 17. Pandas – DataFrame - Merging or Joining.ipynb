```python
# Introduction
# By using pandas we can perform join operations as well.
# This is same as join operations in database.
# Here we are performing join in between the DataFrames
# Joining is the process of bringing two DataFrames into one DataFrame based on common attributes in columns
```

```python
import pandas as pd
```

```python
d1 = {
    "Id": [1, 2, 3, 4, 5, 6],
    "Name": ["Pradhan", "Venu", "Madhurima", "Nireekshan", "Shafi", "Veeru"],
    "Subject": ["English", "Java", "Html", "Python", "C", "dot net"]
}
d2 = {
    "Id": [11, 12, 13, 14, 15, 16],
    "Name": ["Srinu", "Sumanth", "Neelima", "Daniel", "Arjun", "Veeru"],
    "Subject": ["Java", "Html", "Cpp", "Python", "C", "dot net"]
}
```

```python
df1 = pd.DataFrame(d1)
df2 = pd.DataFrame(d2)
```

```python
df1
```

```python
df2
```

```python
inn_join = pd.merge(df1,df2,on = "Subject",how="inner")
```

```python
inn_join
```

```python
# Left join
# Keep every row in the left dataframe.
# Where there are missing values of the "on" variable in the right dataframe filled with NaN values in the result.

d1 = {
    "Id":[1, 2, 3, 4, 5, 6],
    "Name": ["Pradhan", "Venu", "Madhurima", "Nireekshan", "Shafi", "Veeru"],
    "Subject":["English", "Java", "Html", "Python", "C", "dot net"]
}

d2 = {
    "Id":[11, 12, 13, 14, 15, 16],
    "Name": ["Srinu", "Sumanth", "Neelima", "Daniel", "Arjun", "Veeru"],
    "Subject":["Java", "Html", "Cpp", "Python", "C", "dot net"]
}
```

```python
df11 = pd.DataFrame(d1)
df22 = pd.DataFrame(d2)
```

```python
df11
```

```python
df22
```

```python
left_join = pd.merge(df11,df22,on = "Subject", how="left")
```

```python
left_join
```

```python
# Right join

right_join = pd.merge(df11,df22,on = "Subject",how ="right")
```

```python
right_join
```

```python
#  Outer Merge / Full outer join

full_join = pd.merge(df11,df22,on = 'Subject',how = "outer")
```

```python
full_join
```

```python
# Merging two DataFrames based on column
d1 = {
    "Id":[1, 2, 3, 4, 5, 6],
    "Name": ["Pradhan", "Venu", "Madhurima", "Nireekshan", "Shafi", "Veeru"],
    "Subject":["English", "Java", "Html", "Python", "C", "dot net"]
}


d2 = {
    "Id":[11, 12, 13, 14, 15, 16],
    "Name": ["Srinu", "Sumanth", "Neelima", "Daniel", "Arjun", "Veeru"],
    "Subject":["Java", "Html", "Cpp", "Python", "C", "dot net"]
}
```

```python
df1 = pd.DataFrame(d1)
df2 = pd.DataFrame(d2)
```

```python
print(df1)
print()
print("**_**_**_**_**_**_**_**_**")
print()
print(df2)
```

```python
result = pd.merge(df1,df2,on='Subject')
result
```

```python
result = pd.merge(df1,df2,on='Subject',how='right')
result
```

```python
# Merging two DataFrames with multiple columns

result = pd.merge(df1,df2,on=['Name','Subject'])
result
```
