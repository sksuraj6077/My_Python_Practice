```python
# Adding column to DataFrame

import pandas as pd

df = pd.read_csv("sales8.csv")

df.head()
```

```python
df['Status'] = "Delivered"
```

```python
df.head()
```

```python
df["Total Cost"] = df['Product cost'] * df['Quantity']
df.head()
```

```python
df11 = df.sort_values(by="Total Cost",ascending=1)
df11.head()
```

```python
def total(df):
    t = df['Product cost'] * df['Quantity']
    return t
```

```python
df['Total Cost'] = df.apply(total,axis = 1)
```
